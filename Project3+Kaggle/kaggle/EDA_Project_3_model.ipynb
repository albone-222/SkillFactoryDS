```python
import pandas as pd
```

```python
hotels = pd.read_csv('data/hotels.csv')
hotels.head(3)
```

```python
hotels['review_date'] = pd.to_datetime(hotels['review_date'])
```

```python
tags = dict()
for cell in hotels['tags']:
    cell = cell.replace('[', '').replace(']', '')\
        .replace('\' ', '').replace(' \'', '').split(', ')
    for tag in cell:
        if tag in tags.keys():
            tags[tag] += 1
        else:
            tags[tag] = 1
        

sorted_tags = {k: v for k, v in sorted(tags.items(), key=lambda item: item[1], reverse=True)}
print(len(sorted_tags))
sorted_tags
```

```python
def transform_tags(cell):
    cell = cell.replace('[', '').replace(']', '')\
        .replace('\' ', '').replace(' \'', '').split(', ')
    return cell

def get_country(cell):
    cell = cell.split()
    if cell[-1] == 'Kingdom':
        return ' '.join(cell[-2:])
    else:
        return str(*cell[-1:])

def get_city(cell):
    cell = cell.split()
    if cell[-1] == 'Kingdom':
        return str(*cell[-5:-4])
    else:
        return str(*cell[-2:-1])
    


hotels['country'] = hotels['hotel_address'].apply(get_country)
hotels['city'] = hotels['hotel_address'].apply(get_city)
hotels['tags'] = hotels['tags'].apply(transform_tags)
```

```python
hotels['count_tags'] = hotels['tags'].apply(len)
```

```python
hotels['days_since_review'] = hotels['days_since_review'].apply(lambda x: int(x.split()[0]))
hotels
```

```python
hotels['positive_review'].value_counts().head(25)
```

```python
def pos_update(cell):
    replace_dict = {
        'no positive': 'no_positive',
        'location': 'good_location',
        'everything': 'all_good',
        'nothing': 'all_good',
        'breakfast': 'good_brealfast',
        'staff': 'good_staff',
        'bed': 'good_bed'   
    }
    for key in replace_dict.keys():
        if cell.lower().find(key) > 0:
            return replace_dict[key]

    return 'no_positive'
        
hotels['positive_review_update'] = hotels['positive_review'].apply(pos_update)
```

```python
hotels['positive_review_update'].value_counts()
```

```python
hotels['negative_review'].value_counts().head(25)
```

```python
def neg_update(cell):
    replace_dict = {
        'no negative': 'no_negative',
        'location': 'bad_location',
        'everything': 'all_bad',
        'nothing': 'all_bad',
        'breakfast': 'bad_brealfast',
        'price': 'bad_price',
        'room': 'bad_room',
        'none': 'no_negative',
        'n a': 'no_negative',
        'all good': 'no_negative',
        '': 'no_negative',
        'no complaints': 'no_negative',
        'nil': 'no_negative'        
    }
    for key in replace_dict.keys():
        if cell.lower().find(key) > 0:
            return replace_dict[key]

    return 'no_negative'
        
hotels['negative_review_update'] = hotels['negative_review'].apply(neg_update)
```

```python
city_geo = hotels.groupby('city').mean()
city_geo
```

```python
for city in city_geo.index:
    hotels[hotels['city'] == city]['lat'] = hotels['lat'].fillna(city_geo['lat'][city], inplace=True)
    hotels[hotels['city'] == city]['lng'] = hotels['lng'].fillna(city_geo['lng'][city], inplace=True)
    
```

```python
hotels_encoding = hotels.copy()
for tag in list(sorted_tags)[:20]:
    hotels_encoding[tag] = hotels_encoding['tags'].apply(lambda x: 1 if tag in x else 0)
```

```python
hotels_encoding = pd.get_dummies(hotels_encoding, columns=['country', 'city', 'positive_review_update', 'negative_review_update'])
```

```python
hotels_encoding.info()
```

```python
hotels_encoding.drop(columns=['tags', 'hotel_address', 'hotel_name',
                              'reviewer_nationality', 'negative_review', 
                              'positive_review'], inplace=True)
hotels_encoding.info()
```

```python
hotels_encoding.drop(columns=['review_date'], inplace=True)
#hotels_encoding['review_date'] = hotels_encoding['review_date'].values.astype('datetime64[D]')
```

```python
hotels_encoding.info()
```

```python
#    ,        
#  -     ,  -   ( )  
X = hotels_encoding.drop(['reviewer_score'], axis = 1)  
y = hotels_encoding['reviewer_score'] 
```

```python
#     :  
from sklearn.model_selection import train_test_split  
```

```python
#     "train"     , "test" -  .  
#      25%   .  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#   :  
from sklearn.ensemble import RandomForestRegressor #        
from sklearn import metrics #       
  
#    
regr = RandomForestRegressor(n_estimators=100)  
      
#        
regr.fit(X_train, y_train)  
      
#          .  
#      y_pred  
y_pred = regr.predict(X_test)  
```

```python
#    (y_pred)   (y_test),       
#   Mean Absolute Percentage Error (MAPE)          .  
print('MAPE:', metrics.mean_absolute_percentage_error(y_test, y_pred))
```
