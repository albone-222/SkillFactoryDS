```python
import pandas as pd
ufo_witness = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv')
display(ufo_witness)
```

```python
ufo_witness_work = ufo_witness.copy()
ufo_witness_work['Time'] = pd.to_datetime(ufo_witness_work['Time'])
display(ufo_witness_work)
```

```python
often_year_ufo = ufo_witness_work['Time'].dt.year
often_year_ufo.value_counts(normalize = True)
```

```python

date_ufo = ufo_witness_work[ufo_witness_work['State'] == 'NV']['Time'].dt.date.diff().dt.days.mean()
display(date_ufo)
```
